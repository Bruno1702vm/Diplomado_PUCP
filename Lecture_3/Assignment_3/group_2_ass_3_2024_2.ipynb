Import the `REC0111.sav`, `RE223132.sav` and `RE516171.sav` files and their **variables and values labels** from this path `"../../_data/endes/"`. For this excersie you must use a for loop. This loop must iterate over **2019, 2018, 2017, 2016, 2015 folders** and import these files. All the files have the same name. You must store these files and their labels in a nested dictionary named as `all_data`. The keys of the dictionary should be named as `year_2019`, for example, and the keys of the nested dictionary should be `data`, `var_labels`, and `value_labels`. **Hint: Use [this link](https://notebooks.githubusercontent.com/view/ipynb?browser=chrome&color_mode=auto&commit=4d6de78e00e7001f16bf6473c2eb7ce24fb611cd&device=unknown_device&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f616c6578616e6465727175697370652f4469706c6f6d61646f5f505543502f346436646537386530306537303031663136626636343733633265623763653234666236313163642f4c6563747572655f342f4c6563747572655f342e6970796e62&logged_in=true&nwo=alexanderquispe%2FDiplomado_PUCP&path=Lecture_4%2FLecture_4.ipynb&platform=windows&repository_id=427747212&repository_type=Repository&version=95#4.2.)**

```python
# List of years to iterate over
years = [2019, 2018, 2017, 2016, 2015]

# Define the base path where the .sav files are located
base_path = "../../_data/endes/"

# Nested dictionary to store all data and labels
all_data = {}

# Loop to iterate over each year and load the corresponding data
for year in years:
    # Specific path for each year
    year_path = os.path.join(base_path, str(year))
    
    # File names
    files = ["REC0111.sav", "RE223132.sav", "RE516171.sav"]
    var_names = ["rec_1", "rec_2", "rec_3"]
    
    # Dictionary to store data and labels for the current year
    year_data = {"data": {}, "var_labels": {}, "value_labels": {}}
    
    # Load the .sav files and labels for the current year
    for i, file_name in enumerate(files):
        file_path = os.path.join(year_path, file_name)
        df, meta = pyreadstat.read_sav(file_path)
        year_data["data"][var_names[i]] = df
        year_data["var_labels"][var_names[i]] = meta.column_labels
        year_data["value_labels"][var_names[i]] = meta.value_labels
    
    # Store the dictionary of data and labels in all_data with the year as the key
    all_data[f"year_{year}"] = year_data

# Verify the structure of the all_data dictionary
for year, data in all_data.items():
    print(f"Data for {year}:")
    for key, value in data.items():
        print(f"  {key}: {list(value.keys())}")
    print("\n")
```

These attributes should be named as `var_labels` and `value_labels`.  **Hint: Use [this link](https://notebooks.githubusercontent.com/view/ipynb?browser=chrome&color_mode=auto&commit=4d6de78e00e7001f16bf6473c2eb7ce24fb611cd&device=unknown_device&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f616c6578616e6465727175697370652f4469706c6f6d61646f5f505543502f346436646537386530306537303031663136626636343733633265623763653234666236313163642f4c6563747572655f342f4c6563747572655f342e6970796e62&logged_in=true&nwo=alexanderquispe%2FDiplomado_PUCP&path=Lecture_4%2FLecture_4.ipynb&platform=windows&repository_id=427747212&repository_type=Repository&version=95#4.2.3.)**

```python
# Initialization of dictionaries to store variable labels and value labels for each year  
all_var_labels = {}  
all_value_labels = {}  

# Processing the data for each year  
for year_key, year_data in all_data.items():  
    yearly_var_labels = {}  
    yearly_value_labels = {}  

    for rec_key in year_data['var_labels']:  
        var_labels = year_data['var_labels'][rec_key]  
        value_labels = year_data['value_labels'][rec_key]  

        # Ensure var_labels and value_labels are dictionaries  
        if isinstance(var_labels, dict):  
            yearly_var_labels.update(var_labels)  
        else:  
            # If not a dictionary, convert or filter it if necessary  
            var_labels_dict = {item: item for item in var_labels if item is not None}  
            yearly_var_labels.update(var_labels_dict)  

        if isinstance(value_labels, dict):  
            yearly_value_labels.update(value_labels)  
        else:  
            # If not a dictionary, convert or filter it if necessary  
            value_labels_dict = {item: item for item in value_labels if item is not None}  
            yearly_value_labels.update(value_labels_dict)  

    # Store the collected labels in the main dictionaries with the year as the key  
    all_var_labels[year_key] = yearly_var_labels  
    all_value_labels[year_key] = yearly_value_labels  

# Add the dictionaries as new attributes to the DataFrame  
endes_data_2015_2019.var_labels = all_var_labels  
endes_data_2015_2019.value_labels = all_value_labels  

# Verification: Print the attributes to check if they've been added correctly  
print("Variable labels stored in the DataFrame:")  
print(endes_data_2015_2019.var_labels)  

print("\nValue labels stored in the DataFrame:")  
print(endes_data_2015_2019.value_labels) 
```
