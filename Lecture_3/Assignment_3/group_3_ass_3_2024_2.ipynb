```python
import pandas as pd
import pyreadstat

# List of years to iterate over
years = [2019, 2018, 2017, 2016, 2015]

# Initialize the nested dictionary to store all the data
all_data = {}

# Iterate over each year
for year in years:
    # Generate the path for each file according to the year
    base_path = f"../../_data/endes/{year}/"
    
    # Define the file names
    rec1_filename = "REC0111.sav"
    rec2_filename = "RE223132.sav"
    rec3_filename = "RE516171.sav"
    
    # Import each file with its labels
    rec_1, meta1 = pyreadstat.read_sav(base_path + rec1_filename)
    rec_2, meta2 = pyreadstat.read_sav(base_path + rec2_filename)
    rec_3, meta3 = pyreadstat.read_sav(base_path + rec3_filename)
    
    # Extract variable and value labels
    var_labels1 = meta1.column_labels
    value_labels1 = meta1.value_labels
    
    var_labels2 = meta2.column_labels
    value_labels2 = meta2.value_labels
    
    var_labels3 = meta3.column_labels
    value_labels3 = meta3.value_labels
    
    # Save the data and labels in the nested dictionary
    all_data[f'year_{year}'] = {
        'data': {
            'rec1': rec_1,
            'rec2': rec_2,
            'rec3': rec_3
        },
        'var_labels': {
            'rec1': var_labels1,
            'rec2': var_labels2,
            'rec3': var_labels3
        },
        'value_labels': {
            'rec1': value_labels1,
            'rec2': value_labels2,
            'rec3': value_labels3
        }
    }

# Example of accessing the data for a specific year, for example, 2019
print(all_data['year_2019']['data']['rec1'].head())
print(all_data['year_2019']['var_labels']['rec1'])
print(all_data['year_2019']['value_labels']['rec1'])
```

7. Use `all_data` to append all the data sets. Store all data sets in a list using `for loop`. Then, use `pd.concat` to append all the data sets. Also, you must reset the index to have a good-looking data.
```python
import pandas as pd

# Create an empty list to store all the DataFrames
data_list = []

# Iterate over each year in the all_data dictionary
for year in all_data:
    # Extract the DataFrames rec1, rec2, rec3 for the current year
    rec1 = all_data[year]['data']['rec1']
    rec2 = all_data[year]['data']['rec2']
    rec3 = all_data[year]['data']['rec3']
    
    # Add each DataFrame to the list
    data_list.append(rec1)
    data_list.append(rec2)
    data_list.append(rec3)

# Use pd.concat to concatenate all the DataFrames in the list
endes_data_2015_2019 = pd.concat(data_list, ignore_index=True)

# Reset the index for the final DataFrame
endes_data_2015_2019.reset_index(drop=True, inplace=True)

# Display the first records to verify the result
print(endes_data_2015_2019.head())
```
