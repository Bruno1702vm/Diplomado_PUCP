```python
## Step 1:  I convert str variables into int64 using a for loop. 
variables=['V201', 'V613', 'V715', 'V511']
for var in variables:
    endes_data_2015_2019[var]=endes_data_2015_2019[var].fillna('900')
    endes_data_2015_2019[var]=endes_data_2015_2019[var].astype('int64')
    endes_data_2015_2019[var].value_counts(dropna=False)
## Step 2: I verify that now the variables have the correct type

print(endes_data_2015_2019['V201'].value_counts(dropna=False))
print(endes_data_2015_2019['V613'].value_counts(dropna=False))
print(endes_data_2015_2019['V715'].value_counts(dropna=False))
print(endes_data_2015_2019['V511'].value_counts(dropna=False))
```
