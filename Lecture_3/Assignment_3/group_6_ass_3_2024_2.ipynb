This code defines a function that takes a df and a list of columns as input. The function first checks for any missing columns by creating a list (missing_columns) that contains the names of columns specified in the input list but not found in the DataFrame. Then, it then creates a new df (selected_df) that includes only the columns from the input list that are present in the original df. Finally, the function returns both the filtered df and the list of missing columns. This process ensures that the necessary columns are selected and verifies if any expected columns are absent.

```python
#Create a function to select and verify columns 
def check_and_select_columns(df, columns):
    missing_columns = [col for col in columns if col not in df.columns]
    selected_df = df.loc[:, [col for col in columns if col in df.columns]]
    return selected_df, missing_columns
```

This code creates three dictionaries: 'data_dict' associates dataset names with their df, 'metadata_dict' links dataset names to their variable labels, and 'value_labels_dict' connects dataset names to their value labels.

```python
# Dictionary with df and their original labels
data_dict = {
    "rec_1": rec1_1,
    "rec_2": rec2_1,
    "rec_3": rec3_1}

metadata_dict = {
    "rec_1": var_labels1,
    "rec_2": var_labels2,
    "rec_3": var_labels3}

value_labels_dict = {
    "rec_1": value_labels1,
    "rec_2": value_labels2,
    "rec_3": value_labels3}
```

This code creates empty dictionaries 'new_var_labels' and 'new_value_labels' for updated label names. The 'update_labels_for_datasets' function iterates through keys in 'metadata_dict' and 'value_labels_dict', using enumerate to generate new label names. For each key, it checks if the corresponding value is a dictionary; if not, it converts the value into a dictionary. The function then populates 'new_var_labels' and 'new_value_labels' with the updated labels, returning these dictionaries.

```python
# Updated label names' dictionary
new_var_labels = {}
new_value_labels = {}

# Function to update label names
def update_labels_for_datasets(metadata_dict, value_labels_dict):
    new_var_labels = {}
    new_value_labels = {}
    
    for i, key in enumerate(metadata_dict.keys(), start=1):
        # Check if metadata_dict[key] is a dictionary
        if isinstance(metadata_dict[key], dict):
            new_var_labels[f"new_var_labels{i}"] = metadata_dict[key]
        else:
            # If it's not a dictionary, convert it to a dictionary
            new_var_labels[f"new_var_labels{i}"] = {j: val for j, val in enumerate(metadata_dict[key])}
        
        # Check if value_labels_dict[key] is a dictionary
        if isinstance(value_labels_dict[key], dict):
            new_value_labels[f"new_value_labels{i}"] = value_labels_dict[key]
        else:
            # If it's not a dictionary, convert it to a dictionary
            new_value_labels[f"new_value_labels{i}"] = {j: val for j, val in enumerate(value_labels_dict[key])}
    
    return new_var_labels, new_value_labels
```

Finally, this code starts by 'calling the update_labels_for_datasets' function with 'metadata_dict' and 'value_labels_dict' as inputs, which updates and returns 2 dictionaries: 'new_var_labels' and 'new_value_labels'. Then it prints the updated variable labels by iterating over 'new_var_labels'. If the labels are stored in a dictionary, it prints each variable name and its associated label; otherwise, it prints the labels directly. Similarly, it prints the updated value labels from 'new_value_labels', using the same approach to handle both dictionaries and lists. This process ensures that the updated labels are displayed in a readable format.

```python
# Call the function to update the labels
new_var_labels, new_value_labels = update_labels_for_datasets(metadata_dict, value_labels_dict)

# Print results to check
print("New Variable Labels:")
for key, labels in new_var_labels.items():
    print(f"\n{key}:")
    if isinstance(labels, dict):
        for var, label in labels.items():
            print(f"  {var}: {label}")
    else:
        for label in labels:
            print(f"  {label}")

print("\nNew Value Labels:")
for key, labels in new_value_labels.items():
    print(f"\n{key}:")
    if isinstance(labels, dict):
        for var, label in labels.items():
            print(f"  {var}: {label}")
    else:
        for label in labels:
            print(f"  {label}")
```

In this way, the function updated the variable and value dictionaries.

```python
# List of years to process.
years = [2019, 2018, 2017, 2016, 2015]

# Initializing the dictionary that will hold the results
all_data = {}

# List of filenames to be imported
files = ['REC0111.sav', 'RE223132.sav', 'RE516171.sav']

# Looping through each year to load the files and their labels
for year in years:
    # Creating the key for the current year in the dictionary
    year_key = f'year_{year}'
    
    # Setting up the nested dictionary structure for the current year
    all_data[year_key] = {
        'data': {},
        'var_labels': {},
        'value_labels': {}
    }
    
    # Looping through each file to load data and labels
    for file in files:
        # Building the file path
        file_path = os.path.join(f'../../_data/endes/{year}', file)
        
        # Reading the .sav file and its metadata
        data, meta = pyreadstat.read_sav(file_path)
        
        # Extracting the base name (without the .sav extension) for use as the key
        file_key = os.path.splitext(file)[0].lower()
        
        # Storing the data and labels in their respective dictionaries
        all_data[year_key]['data'][file_key] = data
        all_data[year_key]['var_labels'][file_key] = meta.column_names_to_labels
        all_data[year_key]['value_labels'][file_key] = meta.variable_value_labels

# Displaying the resulting dictionary
all_data
```
