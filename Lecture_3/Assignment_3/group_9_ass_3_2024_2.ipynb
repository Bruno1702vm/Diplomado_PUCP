## Question 6

Now, replicate your code of the prevoius sections but for years **2019, 2018, 2017, 2016, 2015**. Import the `REC0111.sav`, `RE223132.sav` and `RE516171.sav` files and their **variables and values labels** from this path `"../../_data/endes/"`. For this excersie you must use a for loop. This loop must iterate over **2019, 2018, 2017, 2016, 2015 folders** and import these files. All the files have the same name. You must store these files and their labels in a nested dictionary named as `all_data`. The keys of the dictionary should be named as `year_2019`, for example, and the keys of the nested dictionary should be `data`, `var_labels`, and `value_labels`. **Hint: Use [this link](https://notebooks.githubusercontent.com/view/ipynb?browser=chrome&color_mode=auto&commit=4d6de78e00e7001f16bf6473c2eb7ce24fb611cd&device=unknown_device&enc_url=68747470733a2f2f7261772e67697468756275736572636f6e74656e742e636f6d2f616c6578616e6465727175697370652f4469706c6f6d61646f5f505543502f346436646537386530306537303031663136626636343733633265623763653234666236313163642f4c6563747572655f342f4c6563747572655f342e6970796e62&logged_in=true&nwo=alexanderquispe%2FDiplomado_PUCP&path=Lecture_4%2FLecture_4.ipynb&platform=windows&repository_id=427747212&repository_type=Repository&version=95#4.2.)**

### Solution

```python
# Creating all_data dictionary
#all_data = {}
#path = f'../../_data/endes/'

all_data = {}
path = f'D:/Documents/GitHub/Diplomado_PUCP/_data/endes'

for year in range(2015, 2019 + 1):
    # Creating nested dictionary for each year
    all_data[f'year_{year}'] = {}

    # Reading sav files and creating variables labels and value variable labels objects 
    path_1 = f'{path}\\{year}\\REC0111.sav'
    rec_1, meta_data = pyreadstat.read_sav(path_1)
    var_labels1 = meta_data.column_names_to_labels
    value_labels1 = meta_data.variable_value_labels

    path_2 = f'{path}\\{year}\\RE223132.sav'
    rec_2, meta_data = pyreadstat.read_sav(path_2)
    var_labels2 = meta_data.column_names_to_labels
    value_labels2 = meta_data.variable_value_labels
    
    path_3 = f'{path}\\{year}\\RE516171.sav'
    rec_3, meta_data = pyreadstat.read_sav(path_3)
    var_labels3 = meta_data.column_names_to_labels
    value_labels3 = meta_data.variable_value_labels

    # Selecting selected columns
    rec1_1 = rec_1.reindex(columns = [col for col in columns_rec1 if col in rec_1.columns])
    rec2_1 = rec_2.reindex(columns = [col for col in columns_rec2 if col in rec_2.columns])
    rec3_1 = rec_3.reindex(columns = [col for col in columns_rec3 if col in rec_3.columns])

    # Selecting keys and variables labels from columns selected
    new_var_labels1 = {key: var_labels1[key] for key in columns_rec1 if key in var_labels1.keys()}
    new_var_labels2 = {key: var_labels2[key] for key in columns_rec2 if key in var_labels2.keys()}
    new_var_labels3 = {key: var_labels3[key] for key in columns_rec3 if key in var_labels3.keys()}

    # Selecting keys and value labels from columns selected
    new_value_labels1 = {key: value_labels1[key] for key in columns_rec1 if key in value_labels1.keys()}
    new_value_labels2 = {key: value_labels2[key] for key in columns_rec2 if key in value_labels2.keys()}
    new_value_labels3 = {key: value_labels3[key] for key in columns_rec3 if key in value_labels3.keys()}

    # By adding the column 'year' for each year
    rec1_1.loc[:, 'year'] = year
    new_var_labels1.update({'year': 'Year of the survey'})

    # Merging rec1_1, rec2_1 and rec3_1
    aux = pd.merge(rec1_1, rec2_1, on='CASEID', how='inner')
    data = pd.merge(aux, rec3_1, on='CASEID', how='inner')

    # Unifying the variable labels dictionaries
    var_labels = new_var_labels1.copy()
    var_labels.update(new_var_labels2)
    var_labels.update(new_var_labels3)

    # Unifying the value labels dictionaries
    value_labels = new_value_labels1.copy()
    value_labels.update(new_value_labels2)
    value_labels.update(new_value_labels3)   

    # Giving attributes to 'data' dataframe
    data.attrs['var_labels'] = var_labels
    data.attrs['value_labels'] = value_labels    

    # By assigning data, var_labels and value_labels to nested dictionary for each year
    all_data[f'year_{year}']['data'] = data.copy()
    all_data[f'year_{year}']['var_labels'] = var_labels.copy()
    all_data[f'year_{year}']['value_labels'] = value_labels.copy()
```

```python
all_var_labels = {}
all_value_labels = {}

# Path, file location
base_path = 'D:/Documents/GitHub/Diplomado_PUCP/_data/endes'

# We define a loop to enter the files according to the years and assign the keys
# Checking the key assignment in the dictionary
for year in range(2015, 2020):
    year_str = str(year)
    all_var_labels[year_str] = {}
    all_value_labels[year_str] = {}
    # verify iteration
    # print(all_var_labels.keys())
    print(all_value_labels.keys())
    
    # Define a dictionary called "files" to assign a tuple and use it as a vehicle to extract information from the different files by year
    files = {
        'REC0111.sav': ('var_labels1', 'value_labels1'),
        'RE223132.sav': ('var_labels2', 'value_labels2'),
        'RE516171.sav': ('var_labels3', 'value_labels3')
    }
    
    for file_name, labels in files.items():
        # assign file location based on previous data and temporary variables inside the loop; read .sav
        file_path = f'{base_path}/{year_str}/{file_name}'
        df, meta_data = pyreadstat.read_sav(file_path)
        
        # extract data, columns are labels and values of variable are values
        var_labels = meta_data.column_names_to_labels
        value_labels = meta_data.variable_value_labels
        
        # copy information and assign it in the variables all_var_labels and all_value_labels, is updated in each iteration
        if labels[0] == 'var_labels1':
            all_var_labels[year_str].update(var_labels)
        elif labels[0] == 'var_labels2':
            all_var_labels[year_str].update(var_labels)
        elif labels[0] == 'var_labels3':
            all_var_labels[year_str].update(var_labels)
        
        if labels[1] == 'value_labels1':
            all_value_labels[year_str].update(value_labels)
        elif labels[1] == 'value_labels2':
            all_value_labels[year_str].update(value_labels)
        elif labels[1] == 'value_labels3':
            all_value_labels[year_str].update(value_labels)
```
