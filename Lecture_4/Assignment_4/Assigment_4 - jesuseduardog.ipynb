# INICIO TAREA: FUNCIÓN

```python
def import_data_sbs_banks(start_month: int, start_year: int, end_month = None, end_year = None):
# end_month = None and end_year = None are the default values 

    if not isinstance( start_month , int ):
        start_month = int(start_month)
    if not isinstance( start_year , int ):
        start_year = int(start_year)
    if end_month is not None and not isinstance(end_month, int):
        end_month = int(end_month)
    if end_year is not None and not isinstance(end_year, int):
        end_year = int(end_year)
        
    # Definir la ruta base
    base_path = '../../_data/sbs/B_RawData/bancos'
    
    # Normalizar los argumentos end_month y end_year si no se proporcionan
    if end_month is None:
        end_month = start_month
    if end_year is None:
        end_year = start_year

    # Crear un diccionario anidado para almacenar los dataframes
    nested_dict = {}
    
    # Iterar sobre los años y meses especificados
    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            # Saltar meses fuera del rango especificado
            if (year == start_year and month < start_month) or (year == end_year and month > end_month):
                continue # 

            # Buscar archivos con el formato 'table_clean_dd_m_yyyy.xlsx'
            for day in range(31, 0, -1):  # Empezar desde el último día del mes
                file_name = f'table_clean_{day}_{month}_{year}.xlsx'
                file_path = os.path.join(base_path, file_name)
                
                # Verificar si el archivo existe
                if os.path.exists(file_path):
                    # Leer el archivo y almacenarlo en el diccionario anidado
                    df = pd.read_excel(file_path)
                    if any(df.columns.str.contains(r"COMERCIAL?", regex=True)): # Si alguna de las columnas contiene COMERCIAL, realizar lo siguiente:
                        df = pd.read_excel(file_path,skiprows=2)  # Saltar las primeras dos filas porque no son informativas
                    if 'Tasa Anual (%)' in df.columns: # Si aparece Tasa Anual (%) como nombre de alguna columna, realizar lo siguiente
                        df = df[~df['Tasa Anual (%)'].str.contains('Nota', na=False)] # retirar la nota al pie de página
                    month_str = str(month).zfill(2) # rellena cadenas con ceros a la izquierda hasta alcanzar la longitud de 2 caracteres 
                    year_str = str(year)
                    if year_str not in nested_dict:
                        nested_dict[year_str] = {}
                    nested_dict[year_str][month_str] = df
                    break  # sale del if 

    return nested_dict
```
