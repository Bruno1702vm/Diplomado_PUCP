Using a loop

```python
import pandas as pd

years = np.arange(2015, 2020)
dfs = []

for year in years:
    file_path = f'../_data/endes/{year}/RECH1.SAV'  # definimos la ruta (que irá cambiando año a año)
    df = pd.read_spss(file_path) # se lee el archivo
    df['year_sample'] = year # se añade una columna con el año
    dfs.append(df) # se coloca cada base debajo de la otra

# Concatenate the DataFrames for different years into a single DataFrame, para cada año es una lista
result_df = pd.concat(dfs, ignore_index=False) 
```

1. We have data `MotherData.csv` excerpted from a recent Demographic and Health Survey.  First convert the dataset from `wide` (each observation is a mother) to `long` (each observation is a birth, with associated mother id). The id `caseid` identifies uniquely all the mothers.  These columns refer to variable of children **['bidx', 'bord', 'b0', 'b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'b7', 'b8', 'b9', 'b10', 'b11', 'b12', 'b13', 'b15', 'b16']** and have new columns for all their children. We have information for 20 children. It starts from last child to oldest one. Use for loops to reshape this dataset from `wide` to `long` ate mother and children level. If you want to get more information from the columns please see [this pdf](http://www.dhsprogram.com/pubs/pdf/DHSG4/Recode6_DHS_22March2013_DHSG4.pdf).

Max number of children is 20. We want data at mother, child level. From children, we only want month (b1) and year (b2) of birth (b4), and sex and have data

```python
df_mother = pd.read_csv('../_data/MotherData.csv')
```

```python
list(df_mother.iterrows()) # es una lista compuesta por muchas tuplas, conteniendo el número de fila y la información de dicha fila
```

```python
i = 8
f'bord_{i:02d}'
# i: Es la variable cuyo valor será insertado.
#:02d: Especifica el formato del número. Aquí:
# 02 indica que el número debe estar compuesto por al menos 2 dígitos, añadiendo ceros a la izquierda si es necesario.
# d indica que el valor es un número entero (dígito).
```

```python
reshaped_data = [] # aquí se guardará la información
# Iterate through each row in the dataframe
for index, row in df_mother.iterrows(): # index es número de fila, row es la información de la columna
# para trabajar con este tipo de loops simultáneo (index, row) es necesaria una lista con tuplas
    caseid = row['caseid']
    # Iterate over the 20 sets of columns
    for i in range(1, 21):
        bord_col = f'bord_{i:02d}' # bord es birthday order (orden de nacimiento)
        b1_col = f'b1_{i:02d}'
        b2_col = f'b2_{i:02d}'
        b4_col = f'b4_{i:02d}'
        
        # Check if bord_col is NaN, and if so, continue to the next iteration
        if pd.isna(row[bord_col]):
            break  # para este loop, ya no revisa más hijos
        
        # Append the reshaped row to the list
        reshaped_data.append({
            'caseid': caseid,
            'bord': row[bord_col],
            'b1': row[b1_col], # mes de nacimiento del hijo
            'b2': row[b2_col], # año de nacimiento del hijo
            'b4': row[b4_col] # sexo del hijo
        })

# Convert the list to a dataframe
reshaped_df_mother = pd.DataFrame(reshaped_data)
reshaped_df_mother
```

```python
reshaped_df_mother.sort_values( ['caseid', 'bord'] ).head(20)
```

```python

# Use pandas wide_to_long function
df_mother_long = pd.wide_to_long(df_mother, stubnames=['bord', 'b1', 'b2', 'b4'], 
                                 i='caseid', j='index', sep='_', suffix='\\d+') 
# bord, b1, b2, b3 y b4 son el comienzo del nombre de las columnas para las que se tiene la información de hijos
# i es la columna que se mantndrá fija
# j es la columna que se crearán con el id
# sep = "_" dice que luego de bord va un guión, y así con las otras columnas
# suffic = '\\d+' significa que van a haber más de un dígito luego de _

df_mother_long
```

```python
# Reset the index to get a flat dataframe
df_mother_long = df_mother_long.reset_index() # el index pasa a ser una columna, case index ahora es columna
df_mother_long
```

```python
# Drop rows where 'bord' is NaN
df_mother_long = df_mother_long.dropna(subset=['bord'])
df_mother_long
```
