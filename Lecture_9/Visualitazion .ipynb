```python
import matplotlib.pyplot as plt
import math

# Assuming 'enaho' DataFrame already exists
# Grouping data by 'year' and 'sector' and counting occurrences
base2 = enaho.groupby(['year', 'sector']).count()

# Defining labels for sectors
labels = ['Agricultura y pesca', 'Minería', 'Manufactura', 'Construcción', 'Comercio', 
          'Transporte y telecomunicaciones', 'Finanzas', 'Servicios']

# Get the unique years in the dataset
years = base2.index.get_level_values('year').unique()

# Determine the number of rows and columns for the subplots
num_years = len(years)
cols = 2  # Set to 2 columns
rows = math.ceil(num_years / cols)  # Calculate the number of rows needed

# Set the figure size based on the number of rows
plt.figure(figsize=(14, 5 * rows))

# Loop through each year to plot individual donut charts
for i, year in enumerate(years):
    plt.subplot(rows, cols, i + 1)
    
    # Get data for the specific year
    year_data = base2.loc[year]
    
    # Plotting the donut chart
    wedges, texts, autotexts = plt.pie(year_data['conglome'], labels=labels,
                                       autopct='%1.1f%%', pctdistance=0.85)
    
    # Add a white circle at the center to make it a donut chart
    center_circle = plt.Circle((0, 0), 0.50, fc='white')
    plt.gca().add_artist(center_circle)
    
    # Set title for each subplot
    plt.title(f'Distribución de la mano de obra contratada por sector económico en {year}')
    
# Adding notes outside the loop to apply to the whole figure
txt = "Elaboración propia - ENAHO (2011-2019)"
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for the year 2019 and specific education level
data_pos = enaho[(enaho['year'] == "2019") & (enaho['educ'] == 'post-grado universitario')]

# Calculate Q1, Q3, and IQR
Q1 = data_pos['l_salario'].quantile(0.25)
Q3 = data_pos['l_salario'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Count the number of outliers
outliers = data_pos[(data_pos['l_salario'] < lower_bound) | (data_pos['l_salario'] > upper_bound)]
num_outliers = outliers.shape[0]

# Plot the boxplot in vertical format
plt.figure(figsize=(6, 8))
box = sns.boxplot(y="l_salario", data=data_pos, palette='pastel')

# Identify the household with the lowest salary
min_salary_row = data_pos[data_pos["l_salario"] == data_pos["l_salario"].min()]
min_salary_value = min_salary_row["l_salario"].values[0]
min_salary_id_hh = min_salary_row["id_hh"].values[0]

# Label the lowest value on the plot
plt.text(
    x=0 + 0.12,
    y=min_salary_value - 0.05,
    s=f'Lowest: {min_salary_id_hh}',
    ha='center',
    va='bottom',
    color='red',
    fontsize=10
)

# Display the number of outliers on the plot
plt.text(
    x=0 + 0.12,
    y=upper_bound + 0.5,  # Adjust position as needed
    s=f'Number of Outliers: {num_outliers}',
    ha='center',
    va='bottom',
    color='blue',
    fontsize=10
)

# Set labels and title
plt.title('Boxplot of Salaries for Post-Graduate Households (2019)')
plt.xlabel('Households')
plt.ylabel('Salary (l_salario)')

# Show the plot
plt.show()

print(f'Total number of outliers: {num_outliers}')
```
